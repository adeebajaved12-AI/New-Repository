**Backend Logic (The API Connection):**

```python
import gradio as gr
from deep_translator import GoogleTranslator

# Translation Function: Jo instructions ke mutabiq input leta hai aur API se process karwata hai
def translate_process(text, src, dest):
    if not text.strip():
        return " Error: Please enter some text to translate."

    # Language Dictionary: Supported languages for the tool
    lang_codes = {
        'English': 'en', 'Urdu': 'ur', 'Arabic': 'ar', 'French': 'fr',
        'German': 'de', 'Spanish': 'es', 'Hindi': 'hi', 'Turkish': 'tr',
        'Chinese': 'zh-CN', 'Japanese': 'ja', 'Russian': 'ru'
    }

    try:
        # Instruction: Source language selection (Auto detect included)
        source = "auto" if src == "Auto Detect" else lang_codes[src]
        target = lang_codes[dest]

        # Calling Translation API (Task 2 & 3)
        translated = GoogleTranslator(source=source, target=target).translate(text)
        return translated
    except Exception as e:
        return f" Connection Issue: Please check your internet connection."
```

**Professional UI & Launch:**

```python
# 3. Professional UI Styling (Custom CSS)
custom_style = """
.gradio-container {
    background: linear-gradient(135deg, #fdfbfb 0%, #ebedee 100%);
    font-family: 'Segoe UI', Tahoma, Geneva, Verdana, sans-serif;
}
#header-title {
    text-align: center;
    color: #1a2a6c;
    font-weight: 800;
    margin-bottom: 5px;
}
.sub-text {
    text-align: center;
    color: #555;
    margin-bottom: 25px;
}
#translate-btn {
    background: linear-gradient(to right, #1a2a6c, #b21f1f, #fdbb2d) !important;
    color: white !important;
    border: none !important;
    padding: 10px !important;
    font-size: 1.2em !important;
    font-weight: bold !important;
    border-radius: 8px !important;
    box-shadow: 0 4px 6px rgba(0,0,0,0.1) !important;
    cursor: pointer;
}
.input-box textarea, .output-box textarea {
    border: 2px solid #1a2a6c !important;
    border-radius: 10px !important;
}
"""

# UI Layout Design
languages = sorted(['English', 'Urdu', 'Arabic', 'French', 'German', 'Spanish', 'Hindi', 'Turkish', 'Chinese', 'Japanese', 'Russian'])

with gr.Blocks(theme=gr.themes.Default(), css=custom_style) as demo:
    with gr.Column():
        gr.HTML("<h1 id='header-title'>🌐 AI Multilingual Translator</h1>")
        gr.HTML("<p class='sub-text'>Internship Project | Developed by <b>Adeeba Malik</b></p>")

        with gr.Row():
            source_dropdown = gr.Dropdown(choices=["Auto Detect"] + languages, value="Auto Detect", label="From:")
            target_dropdown = gr.Dropdown(choices=languages, value="Urdu", label="To:")

        with gr.Row():
            # Instruction: User interface for text input
            input_text = gr.Textbox(placeholder="Enter your content here...", label="Original Text", lines=6, elem_classes=["input-box"])
            # Instruction: Display translated text clearly with copy button
            output_text = gr.Textbox(label="Translated Result", lines=6, show_copy_button=True, elem_classes=["output-box"])

        translate_btn = gr.Button("TRANSLATE NOW", elem_id="translate-btn")

        gr.Markdown("---")
        gr.HTML("""
            <div style='text-align: center; color: #666; font-size: 0.85em;'>
                Official AI Project | Faculty of Computing & Engineering <br>
                University of Kotli Azad Jammu and Kashmir
            </div>
        """)

    # Connect the UI to the Logic
    translate_btn.click(
        fn=translate_process,
        inputs=[input_text, source_dropdown, target_dropdown],
        outputs=output_text
    )

# share=True provides a public URL without password
demo.launch(share=True)
```
